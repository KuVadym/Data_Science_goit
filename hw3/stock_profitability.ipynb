```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = {'Момент наблюдения': [1, 2, 3, 4, 5, 6], 'Доходность A, %': [25, -10, 10, 5, 35, 13], 'Доходность B, %': [0, 15, -5, 5, 20, 25], 'Доходность C, %': [10, 25, -15, -5, -5, 15]}
df = pd.DataFrame(data)
df
```

```python
sns.lmplot(x="Доходность A, %", y="Доходность B, %", data=df, lowess=False)
```

```python
sns.lmplot(x="Доходность A, %", y="Доходность C, %", data=df)
```

```python
sns.lmplot(x="Доходность B, %", y="Доходность C, %", data=df)
```
