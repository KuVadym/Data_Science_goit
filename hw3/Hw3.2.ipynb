```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import randrange


fig = plt.figure()
ax = fig.add_subplot()

x = [1, 50, 100]
y = [1, 50, 1]

new_x_el = 25
new_y_el = 10

for i in range(700):
    point = randrange(3)
    if x[point] < new_x_el: 
        new_x_el = (new_x_el+x[point])/2
    if x[point] > new_x_el:
        new_x_el = (x[point]+new_x_el)/2
    if y[point] < new_y_el: 
        new_y_el = (new_y_el+y[point])/2
    if y[point] > new_y_el:
        new_y_el = (y[point]+new_y_el)/2

    x.append(new_x_el)
    y.append(new_y_el)


ax.scatter(x, y)

plt.show()
```
