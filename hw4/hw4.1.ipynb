```python
from scipy import integrate
import numpy as np
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, num=12)
time
```

```python
import matplotlib.pyplot as plt
plt.plot(time, speed)
plt.axis([0, 11, 0, 130])
plt.grid(True)
```

```python
from scipy.interpolate import interp1d

f = interp1d(time, speed, kind='cubic')

plt.plot(time, speed, 'o')
more_time = np.linspace(0, 11, 10000)
plt.plot(more_time, f(more_time))
```

```python
result = integrate.quad(f, 0, 11)
print(result)
```

```python
f = interp1d(time, speed, kind='quadratic')

plt.plot(time, speed, 'o')
more_time = np.linspace(0, 11, 10000)
plt.plot(more_time, f(more_time))
```

```python
result = integrate.quad(f, 0, 11)
print(result)
```
